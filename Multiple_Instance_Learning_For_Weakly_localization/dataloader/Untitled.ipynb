```python
import os
import pandas as pd
```

```python
root_path="../../../../../Data/ImagenetLocalization/"
val_name = root_path + "LOC_val_solution.csv"
train_name = root_path + "LOC_train_solution.csv"
map_name = root_path + "LOC_synset_mapping.txt"
```

```python
pd.read_csv(val_name).head(5)
```

```python
# get class name
class_list = []
with open(map_name, "r") as f:
    for line in f:
        class_list.append(line.split(" ")[0])
```

```python
new_df = pd.read_csv(val_name)
```

```python
a = "[1,2,3]"
a.replace([",", "["], "")
```

```python
new_df = pd.read_csv(val_name)
new_df["label"] = new_df["PredictionString"].apply(lambda r: class_list.index(r.split(" ")[0]))
new_df["bbox"] = new_df["PredictionString"].apply(lambda r: str(r.split(" ")[1:])[1:-1].replace(",", "").replace("'", "")[:-1])
new_df = new_df[new_df["bbox"].apply(lambda r: len(r.split(" ")) == 4)]
new_df["type"] = "val"
new_df["path"] = new_df["ImageId"].apply(lambda r: root_path + "ILSVRC/Data/CLS-LOC/val/" + r + ".JPEG")
new_df = new_df.reset_index()
new_df = new_df.drop("index", axis = 1)
new_df = new_df.drop("PredictionString", axis = 1)
new_df.head(5)
```

```python
new_df2 = pd.read_csv(train_name)
new_df2["label"] = new_df2["PredictionString"].apply(lambda r: class_list.index(r.split(" ")[0]))
new_df2["bbox"] = new_df2["PredictionString"].apply(lambda r: str(r.split(" ")[1:])[1:-1].replace(",", "").replace("'", "")[:-1])
new_df2 = new_df2[new_df2["bbox"].apply(lambda r: len(r.split(" ")) == 4)]
new_df2["type"] = "train"
new_df2["path"] = new_df2["ImageId"].apply(lambda r: root_path + "ILSVRC/Data/CLS-LOC/train/" + r.split("_")[0] + "/" + r + ".JPEG")
new_df2 = new_df2.reset_index()
new_df2 = new_df2.drop("index", axis = 1)
new_df2 = new_df2.drop("PredictionString", axis = 1)
new_df2.head(5)
```

```python
result = pd.concat([new_df, new_df2])
result.head(5)
```

```python
result.to_csv("localization_data.csv", index = False)
```
